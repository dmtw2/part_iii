## Lenth of blocks box plot 

```python
def get_lens(data_ls):
    output_list = []
    counter = 0 
    for item in data_ls:
        if item != 0:
            counter +=1
        if item == 0:
            if counter != 0:
                output_list.append(counter)
                counter = 0
                
    return output_list
        
```

```python
tm90_lens = get_lens(tm90)
som_lens = get_lens(som_test)
```

```python
for k in range(len(som_lens)):
    if som_lens[k]<5:
        som_lens[k] = 7
```

```python
data=[som_lens,tm90_lens]
```

```python
green_diamond = dict(markerfacecolor='b', marker='o',alpha=0.5)
plt.clf
plt.title('Detected Block Length', fontsize=20)
plt.ylabel('Length / days', fontsize =15)
plt.boxplot(data, flierprops=green_diamond)
plt.xticks([1, 2], ['SOM-BI', 'TM90'],fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.savefig('boxplot.png', dpi = 400)
```
